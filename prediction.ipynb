```python
import pandas as pd
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

### Load ANN Trained Model

```python
model=load_model('model.h5')
with open('onehot_encoder_geo.pkl','rb') as file:
    label_encoder_geo=pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
input_data={
    'CreditScore':900,
    'Geography':'Germany',
    'Gender':'Female',
    'Age': 29,
    'Tenure':4,
    'Balance':110406.74,
    'NumOfProducts':7,
    'HasCrCard':1,
    'IsActiveMember':0,
    'EstimatedSalary':110346.88
}
```

```python
input_data['Gender']=label_encoder_gender.transform([input_data['Gender']])[0]
```

```python
input_geo= pd.DataFrame(label_encoder_geo.transform([[input_data['Geography']]]).toarray(),columns=label_encoder_geo.get_feature_names_out(['Geography']))
```

```python
input_data
```

```python
input_data=pd.DataFrame([input_data])
```

```python
input_data=pd.concat([input_data.drop(columns=['Geography'],axis=1),input_geo],axis=1)
```

```python
input_data
```

```python
input_scaled=scaler.transform(input_data)
```

```python
prediction=model.predict(input_scaled)
```

```python
prediction_probabilty=prediction[0][0]
prediction_probabilty
```

```python
if(prediction_probabilty>0.5):
    print('He is Likely to Churn')
else:
    print('He is NOT Likely to Churn')
```
